```python
import pandas as pd
import plotly.express as px
from datetime import datetime
```

```python
# Load the dataset directly from URL (Best for Colab)
url = "https://raw.githubusercontent.com/selva86/datasets/master/supermarket_sales.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
# 3. Key Metrics (KPIs)

print(" 3. KEY METRICS (KPIs)")
total_revenue = df['Total'].sum()
print(f"Total Revenue     : ${total_revenue:,.2f}")
print(f"Total Transactions: {len(df):,}")
print(f"Avg Order Value   : ${total_revenue/len(df):,.2f}")
print(f"Total Units Sold  : {df['Quantity'].sum():,}")
print(f"Avg Customer Rating: {df['Rating'].mean():.2f}/10")
```

```python
# 2. Sales by Product Line
fig2 = px.bar(df.groupby('Product line')['Total'].sum().sort_values(ascending=False).reset_index(),
              x='Total', y='Product line', orientation='h',
              title='🏷️ Sales by Product Line', color='Total')
fig2.show()
```

```python
# 3. Sales by City
fig3 = px.bar(df.groupby('City')['Total'].sum().reset_index(),
              x='City', y='Total', title='🌆 Sales by City', color='City')
fig3.show()
```

```python
# 4. Product Line Pie Chart
fig4 = px.pie(df, names='Product line', values='Total',
              title='🥧 Revenue Share by Product Line', hole=0.4)
fig4.show()
```

```python
# 5. Payment Methods
fig5 = px.pie(df, names='Payment', values='Total',
              title='💳 Sales by Payment Method')
fig5.show()
```

```python
# 6. Gender & Customer Type
fig6 = px.bar(df.groupby(['Gender', 'Customer type'])['Total'].sum().reset_index(),
              x='Gender', y='Total', color='Customer type', barmode='group',
              title='👥 Sales by Gender and Customer Type')
fig6.show()
```
